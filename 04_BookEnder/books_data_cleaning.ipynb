```python
import gensim
```

```python
model = gensim.models.Word2Vec.load_word2vec_format('./Projects/4Fletcher/GoogleNews-vectors-negative300.bin', binary=True) 
```

```python
input1 = 'i have some apples and i like eat apples'
input2 = 'computer science is great'
input3 = 'i love oranges good to eat'
```

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

output1 = [t for t in word_tokenize(input1) if t not in stopwords.words('english')]
output2 = [t for t in word_tokenize(input2) if t not in stopwords.words('english')]
output3 = [t for t in word_tokenize(input3) if t not in stopwords.words('english')]
```

```python
model_out1 = [model[w] for w in output1]
model_out2 = [model[w] for w in output2]
model_out3 = [model[w] for w in output3]
```

```python
mean_vect1 = np.array(model_out1).mean(axis=0)
mean_vect2 = np.array(model_out2).mean(axis=0)
mean_vect3 = np.array(model_out3).mean(axis=0)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
cosine_similarity(mean_vect1, mean_vect3)
```
