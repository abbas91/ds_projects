```python
import pandas as pd
import numpy as np
import pickle
```

```python
with open('book_vectors.pkl','rb') as f:
    book_df = pickle.load(f)
```

```python
book_df.rename(columns={0:'vectors'}, inplace=True)
book_df.shape
```

```python
book_df.head()
```

```python
# Taking out the Null Values
index = []
vectors = []
for i in range(len(book_df)):
    try:
        if book_df.vectors[i][0]:
            index.append(i)
            vectors.append(book_df.vectors[i])
    except:
        pass
```

```python
len(index)
len(vectors)
# len(vectors[0])
```

```python
# Making a dataframe are re-indexing based on which Nulls were taken out
vectors_df = pd.DataFrame(vectors)
vectors_df['ind'] = index
vectors_df = vectors_df.set_index('ind',drop=True)
```

```python
# Now drop the np.NaN values, this is different from the null values
vectors_df_drop = vectors_df.dropna(how='any', axis=0)
```

```python
vectors_df_drop.head()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Compute cosine similarity
dists = cosine_similarity(vectors_df_drop)
```

```python
# Make DataFrame and re-index based on the book number
dists = pd.DataFrame(dists, columns=vectors_df_drop.index)
dists.set_index(vectors_df_drop.index, inplace=True)
```
